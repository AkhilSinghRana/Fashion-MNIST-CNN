# Feature Engineering


```python
#Import required packages
import os
import pandas as pd
from skimage import io
from skimage.color import rgb2grey
from PIL import Image
import numpy as np

import cv2
# Matplotlib for plotting
from matplotlib import pyplot as plt
#For plotting inline in jupyter-lab
%matplotlib inline 
```

### Define a function that reads an image and return a gray scale image of it

```python
def convertToGray(image_dir, image_filename, saveGray=False, visualize=False, module="skimage"):
    # create a path to the image
    image = os.path.join(image_dir, image_filename)
    
    #skimage
    original_image = io.imread(image) # image is stored just for visualization purposes
    gray_image = rgb2grey(original_image) # Skimage gives the gray scale in the range of [0,1]
    
    #opencv
    opencv_original_image = cv2.imread(image)
    opencv_gray_image = cv2.cvtColor(opencv_original_image, cv2.COLOR_BGR2GRAY)
    
    #Numpy
    np_image = np.asarray(original_image)
    # Using mean Pixel value of each channel(RGB)
    gray_numpy_image = np.mean(np_image, axis=2).astype("uint8")
    
    
    #Let's try a different formula, for Luminance corection, 
        #Y = 0.21 R + 0.72 G + 0.07 B.
        #      Formula source--> https://www.johndcook.com/blog/2009/08/24/algorithms-convert-color-grayscale/
    gray_luminanceCorrected_image = np.dot(np_image[...,:3], [0.21, 0.71, 0.07])
    
    if visualize:
        fig, axes = plt.subplots(1, 8, figsize=(18, 12))
        ax = axes.ravel()

        ax[0].imshow(original_image)
        ax[0].set_title("Original")
        
        ax[1].imshow(gray_image, cmap=plt.cm.gray)
        ax[1].set_title("Grayscale-SkImage")
        
        ax[2].imshow(opencv_gray_image, cmap=plt.cm.gray)
        ax[2].set_title("Grayscale-OpenCV")
        
        ax[3].imshow(gray_numpy_image, cmap=plt.cm.gray)
        ax[3].set_title("Avergare Channels")
        
        
        ax[4].imshow(gray_luminanceCorrected_image, cmap=plt.cm.gray)
        ax[4].set_title("Luminance Corrected")

        ax[5].imshow(original_image[:,:,0])
        ax[5].set_title("R- Channel")
        ax[6].imshow(original_image[:,:,1])
        ax[6].set_title("G-Channel")
        ax[7].imshow(original_image[:,:,2])
        ax[7].set_title("B:Channel")
        fig.tight_layout()
        plt.show()
        
        
    if saveGray:
        output_dir = os.path.join(os.path.dirname(os.path.dirname(image)), "GrayImages")
        if not os.path.isdir(output_dir):
            os.mkdir(output_dir)
        io.imsave(os.path.join(output_dir, "{}.png".format(image_filename)), gray_numpy_image)
    
    
    return gray_numpy_image
```

```python
# Read the images and pass it to the function to get gray scale
data = pd.read_csv("../data/gicsd_labels.csv")
image_dir = "../data/images/"
for i, j in data.iterrows():
    if i == 40:
        print(j[1])
        gray = convertToGray(image_dir, j[0], visualize=True, saveGray=False)
        print("Converted Image's shape is --> ",gray.shape)
        break
```

### Let's modify our gray scale conversion function a bit, to keep most of the information from channel, B. We can think of just saving channel B of the image, but to be on a safe side, it might be possible that R and G channels also hold some information that might be useful for the model

```python
def convertToGray(image_dir, image_filename, saveGray=False, visualize=False, module="skimage"):
    # create a path to the image
    image = os.path.join(image_dir, image_filename)
    
    #skimage
    original_image = io.imread(image) # image is stored just for visualization purposes
    
    
    #Numpy
    np_image = np.asarray(original_image)
    # Using mean Pixel value of each channel(RGB)
    gray_numpy_image = np.mean(np_image, axis=2).astype("uint8")
    
    
    # Now Let's try a different formula, to keep most out of the Blue-Channel, 
        #Y = 0.04 R + 0.18 G + 0.98 B.
    gray_Corrected_image = np.dot(np_image[...,:3], [0.04, 0.18, 0.98])
    
    if visualize:
        fig, axes = plt.subplots(1, 6, figsize=(18, 12))
        ax = axes.ravel()

        ax[0].imshow(original_image)
        ax[0].set_title("Original")
        
        ax[1].imshow(original_image[:,:,0])
        ax[1].set_title("R- Channel")
        ax[2].imshow(original_image[:,:,1])
        ax[2].set_title("G-Channel")
        ax[3].imshow(original_image[:,:,2])
        ax[3].set_title("B:Channel")
        
        ax[4].imshow(gray_numpy_image, cmap=plt.cm.gray)
        ax[4].set_title("Avergare Channels")
        
        
        ax[5].imshow(gray_Corrected_image, cmap=plt.cm.gray)
        ax[5].set_title("B as most important Channel")
        
        

        
        fig.tight_layout()
        plt.show()
        
        
    if saveGray:
        output_dir = os.path.join(os.path.dirname(os.path.dirname(image)), "GrayCorrectedImages")
        if not os.path.isdir(output_dir):
            os.mkdir(output_dir)
        io.imsave(os.path.join(output_dir, "{}".format(image_filename)), gray_Corrected_image)
    
    
    return gray_Corrected_image
```

```python
# Read the images and pass it to the function to get gray scale
data = pd.read_csv("../data/gicsd_labels.csv")
image_dir = "../data/images/"
for i, j in data.iterrows():
    if i == 90:
        print(j[1])
        gray = convertToGray(image_dir, j[0], visualize=True, saveGray=False)
        print("Converted Image's shape is --> ",gray.shape)
        break
```

```python
for i, j in data.iterrows():
    gray = convertToGray(image_dir, j[0], visualize=False, saveGray=True)
print("images Saved")
```
